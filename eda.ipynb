## streotype 가설 확인 
### 중장년 -> 트로트 이런식으로 매핑했는지 등에 대한 확인 

```python
def compare(text_col, keyword, target_col, topn=6):
    sub = df[df[text_col].astype(str).str.contains(keyword, na=False)]
    if len(sub) < 50:
        print(f"[{keyword}] 표본 부족 (n={len(sub)})")
        return
    c = pd.DataFrame({
        f"'{keyword}'언급": sub[target_col].value_counts(normalize=True),
        "전체": df[target_col].value_counts(normalize=True)
    })
    c["차이"] = (c.iloc[:, 0] - c["전체"])
    print(f"\n=== '{keyword}' in {text_col} → {target_col} (n={len(sub):,}) ===")
    print(c.sort_values("차이", ascending=False).head(topn).round(3))

# 첫 3행에서 보인 가설들
compare("arts_persona", "트로트|임영웅|미스터트롯|이찬원", "age")
compare("sports_persona", "등산|산을|산행|산자락", "province")
compare("cultural_background", "갯벌|바닷바람|앞바다", "province")
```

```python
compare("culinary_persona", "막걸리", "age")
compare("hobbies_and_interests", "독서|책", "education_level")
compare("travel_persona", "온천", "age")
compare("professional_persona", "은퇴|정년", "age")
```

## 연령대 추가 확인 

```python
df["age_bucket"] = pd.cut(df["age"], bins=[18, 29, 39, 49, 59, 69, 100],
                          labels=["20대", "30대", "40대", "50대", "60대", "70+"])

def compare_bucket(text_col, keyword, target_col="age_bucket"):
    sub = df[df[text_col].astype(str).str.contains(keyword, na=False)]
    c = pd.DataFrame({
        "언급자": sub[target_col].value_counts(normalize=True),
        "전체": df[target_col].value_counts(normalize=True)
    })
    c["차이"] = c["언급자"] - c["전체"]
    c["배수"] = (c["언급자"] / c["전체"]).round(2)   # 전체 대비 몇 배
    print(f"\n=== '{keyword}' (n={len(sub):,}) ===")
    print(c.sort_values("차이", ascending=False).round(3))

compare_bucket("arts_persona", "트로트|임영웅|미스터트롯|이찬원")
compare_bucket("culinary_persona", "막걸리")
compare_bucket("travel_persona", "온천")
compare_bucket("professional_persona", "은퇴|정년")
```

## 통계적 유의성 검정 

```python
from scipy.stats import chi2_contingency

def chi2_test(text_col, keyword, target_col):
    df["_has_kw"] = df[text_col].astype(str).str.contains(keyword, na=False)
    ct = pd.crosstab(df["_has_kw"], df[target_col])
    chi2, p, dof, _ = chi2_contingency(ct)
    n = df["_has_kw"].sum()
    print(f"'{keyword}' → {target_col}: chi2={chi2:.0f}, "
          f"p={p:.2e}, n={n:,}  {'★유의' if p < 0.001 else ''}")

chi2_test("arts_persona", "트로트|임영웅|미스터트롯|이찬원", "age_bucket")
chi2_test("sports_persona", "등산|산을|산행|산자락", "province")
chi2_test("cultural_background", "갯벌|바닷바람|앞바다", "province")
chi2_test("culinary_persona", "막걸리", "age_bucket")
```

## 역방향검증 
### 키워드 있는 사람의 demographic을 봤는데, 특정 demographic의 키워드 등장률을 확인한다. 

```python
# 광역시도별 '갯벌/바다' 언급률 (지역 stereotype을 가장 명확히 보여줌)
df["_sea"] = df["cultural_background"].astype(str).str.contains("갯벌|바닷바람|앞바다", na=False)
rate = df.groupby("province")["_sea"].mean().sort_values(ascending=False)
print("=== 광역시도별 '바다' 언급률 ===")
print((rate * 100).round(1))

# 연령대별 '트로트' 언급률
df["_trot"] = df["arts_persona"].astype(str).str.contains("트로트|임영웅|미스터트롯|이찬원", na=False)
rate2 = df.groupby("age_bucket")["_trot"].mean()
print("\n=== 연령대별 '트로트' 언급률 ===")
print((rate2 * 100).round(1))
```

```python
# '갯벌/바다' 언급자를 전부 '전라남'으로 찍는 단순 규칙의 정확도
sea_users = df[df["_sea"]]
naive_acc = (sea_users["province"] == "전라남").mean()
print(f"'바다 언급 → 전라남' 단순 규칙 정확도: {naive_acc:.1%} "
      f"(전체 전라남 비율 3.5% 대비)")
```

## 데이터 leakage 이슈에 대한 추가 분석 

```python
import re

# 17개 광역시도 + 주요 시군구 키워드 (간단 버전)
loc_words = ["서울","경기","인천","부산","대구","광주","대전","울산","세종",
             "강원","충청","전라","경상","제주","전남","전북","경남","경북","충남","충북",
             "무안","여수","안양","수원","창원"]  # 필요시 district에서 추출해 확장
loc_pattern = "|".join(loc_words)

# 마스킹 함수
def mask_loc(text):
    return re.sub(loc_pattern, "[LOC]", str(text))

# professional_persona에서 지역명 마스킹 후, '바다' 언급자의 지역 분포가 여전히 전라남에 쏠리는지
df["_prof_masked"] = df["professional_persona"].apply(mask_loc)
df["_sea2"] = df["cultural_background"].astype(str).str.contains("갯벌|바닷바람|앞바다", na=False)

sea_masked = df[df["_sea2"]]
print("마스킹과 무관하게, '바다 언급(cultural_background)' → province 분포:")
print(sea_masked["province"].value_counts(normalize=True).head(5).round(3))
print("\n→ cultural_background의 '바다' 같은 간접 단서만으로도 지역이 쏠린다면,")
print("  이건 지역명 복사가 아니라 진짜 stereotype 학습 신호 (우리가 원하는 것)")
```
